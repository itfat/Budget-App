```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, shuffle= False)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
for column_name in X_train.columns:
    if X_train[column_name].dtype == object:
         X_train[column_name] = le.fit_transform(X_train[column_name])
    else:
        pass
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
# get the model
def get_model(n_inputs, n_outputs):
    model = keras.Sequential()
    model.add(layers.Dense(40, input_dim=n_inputs, kernel_initializer='he_uniform', activation='relu'))
    model.add(layers.Dense(20, input_dim=n_inputs, kernel_initializer='he_uniform', activation='relu'))
    model.add(layers.Dense(n_outputs, kernel_initializer='he_uniform'))
    model.compile(loss='mae', optimizer='adam')
    return model
```

```python
n_inputs, n_outputs = X_train.shape[1], Y_train.shape[1]
```

```python
print(n_inputs, n_outputs)
```

```python
model = get_model(n_inputs, n_outputs)
```

```python
model.fit(X_train, Y_train, verbose=0, epochs=100)
```
